```python
indicadores = df.Indicadores.unique()
```

```python
df['Tipo de indicador'] = None
```

```python
correspondencias = {
    'Diabetes': [
        'diabetes mellitus'
    ],
    'Hepática': [
        'higado',
        'cirrosis',
        'hepatitis',
        'alcohol',
        'Esteatosis hepática',
        'Biopsia hepática',
        'Colangitis',
        'fibrosis hepatica',
        'cannabis',
        'alcohol'
    ],
    'Cardiaca': [
        'cardiopatia',
        'miocardio',
        'cardiaca',
        'arterial',
        'hipertension ',
        'Angioplastia',
        'Arritmia',
        'isquemica',
        'miocardio',
        'hipertensiva'
    ],
    'Pulmonar': [
        'enfermedad obstructuva cronica',
        'EPOC',
        'asma',
        'fibrosis pulmonar',
        'neumonia',
        'influenza',
        'tuberculosis',
        'bronquitis',
        'tabaco',
        'tuberculosis',
        'pulmonar'
    ],
    'Cerebral': [
        'cerebro',
        'cerebrovasculares',
        'Hemorragia cerebral',
        'Isquemia cerebral',
        'Infarto cerebral',
        'cerebrovascular',
        'Tomografía cerebral',
        'Meningitis',
        'Encefalopatía',
        'Sinapsis'
    ],
    'Mental': [
        'salud mental',
        'trastorno mental',
        'Psicología',
        'Psiquiatría',
        'depresión',
        'Ansiedad',
        'Esquizofrenia',
        'Terapia cognitivo-conductual',
        'Trastorno bipolar',
        'suicidio'
    ],
    'Cáncer': 'cancer',
    'Hospitalizaciones':['hostilaraio','hopitalizacion']}
```

```python
#CREAR LA COLUMNA TIPO DE INDICADORES:

def asignar_tipo(row):
    for tipo, indicadores in correspondencias.items():
        if isinstance(indicadores, str):
            if indicadores.lower() in row['Indicadores'].lower():
                return tipo
        else:
            if any(indicador.lower() in row['Indicadores'].lower() for indicador in indicadores):
                return tipo

    return 'Sin tipo'  # Devuelve None si no se encuentra coincidencia

df['Tipo de indicador'] = df.apply(asignar_tipo, axis=1)
df.head()
```

```python
df.to_csv('../../data/processed/Indicadores.csv',index=False)
```
